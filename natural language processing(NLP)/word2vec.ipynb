# 基础用法

```python
from gensim.test.utils import common_texts
from gensim.models import Word2Vec

#common_texts 可以是iterable
#size: 词向量维度
#window: 上下文半宽度
#min_count: 词最小出现次数
#workers: 线程数
model = Word2Vec(common_texts, size=5, window=5, min_count=1, workers=4)
```

```python
common_texts
```

```python
model.save("word2vec.model") #存成二进制文件
model.wv.save_word2vec_format("word2vec_model.txt") #存成与C兼容的txt格式
```

```python
! cat word2vec_model.txt
```

## 将文本转为分词后的文本

```python
with open('corpus.txt','r') as f:
    lines = f.readlines()
    
print(lines)
```

```python
import jieba

line_segs = [list(jieba.cut(l)) for l in lines]
print(line_segs)

line_seg_str = [" ".join(l) for l in line_segs]
print(line_seg_str)
```

```python
with open('corpus_segs.txt','w') as f:
    f.writelines(line_seg_str)
    
! cat corpus_segs.txt
```

## 通过文本加载训练

```python
from gensim.models.word2vec import LineSentence

sentences = LineSentence("corpus_segs.txt")
print(next(iter(sentences)))
```

```python
import multiprocessing
model = Word2Vec(sentences,
                 size=5,
                 window=5,
                 min_count=1,
                 workers=multiprocessing.cpu_count())

model.wv.save_word2vec_format("word2vec_model.txt") #存成与C兼容的txt格式
print(multiprocessing.cpu_count())
```
